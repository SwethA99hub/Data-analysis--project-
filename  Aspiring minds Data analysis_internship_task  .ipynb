####Test claim.

Null Hypothesis ,Is fresher's who work as programming analyst,Hardware engineer,software engineer ,Associate engineers earn 2.5LPA.

Alternative Hypothesis , not Null Hypothesis


```python
from scipy import stats
from scipy.stats import chi2_contingency as cst
```

```python
pv=stats.ttest_1samp(DF['Salary'],popmean=250000)[1]

if pv<0.05:
    print('we reject the null hyothesis and average salary is 2.5lpa')
else:
    print('we fail to reject null hyothesis and average salary is not equal to 2.5lpa')
```

## Relation between Gender & specialization

```python
plt.figure(figsize=(8,4),dpi=200)
sns.countplot(data=DF,hue='Gender',x='Specialization')
plt.show()
```

```python
sample_columns = pd.crosstab(DF['Gender'],DF['Specialization'],margins= True)
pv = cst(sample_columns)[1].any()
if pv<0.05:
    print('we reject the null hypothesis and gender impact specialization')
else:
    print('we fail to reject the null hyothesis')
```
